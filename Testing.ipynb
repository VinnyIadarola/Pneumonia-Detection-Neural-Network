```python
import requests
import numpy as np
import pandas as pd
from io import BytesIO
from PIL import Image, ImageOps

import torch
import torch.nn as nn
```

```python
def preProcess(imgPath):
    img_pil = Image.open(imgPath)
    img_pil = img_pil.resize((600, 500))
    img_pil = ImageOps.grayscale(img_pil)
    img = torch.from_numpy(np.array(img_pil)).float() / 255.
    
    return img
```
